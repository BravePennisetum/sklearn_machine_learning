### 1.加载数据

```python
import pandas as pd
from sklearn.datasets import load_iris
```

加载sklearn数据集有两种方法,分别是load和fetch.load是获取小规模的数据集,包含在datasets中,fetch获取大规模数据集,需要从网络上下载

```python
iris = load_iris()
```
