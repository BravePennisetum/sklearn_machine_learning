特征的预处理主要指特征的无量纲化,包括:  
(1)归一化  
(2)标准化

```python
import pandas as pd
```

```python
data = pd.read_csv('./data/dating.txt')
```

```python
data.head()
```

```python
data = data.iloc[:,0:3]
```

```python
data.head()
```

### 1.归一化(用得少)  
归一化的缺点是易受异常值的影响,因为归一化用的方法是依据最大最小值,而异常值通常就是最大值或最小值

  

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# 1.实例化一个转换器类
transfer = MinMaxScaler()

# 2.调用fit_transform()
data_nor = transfer.fit_transform(data)
```

```python
data_nor
```

### 2.标准化(用的多)  
因为标准化的均值和标准差都是对一堆数求的统计量,所以受异常值的影响较小

```python
from sklearn.preprocessing import StandardScaler
```

```python
# 1.实例化一个转换器类
transfer = StandardScaler()

# 2.调用fit_transform()
data_nor = transfer.fit_transform(data)
```

```python
data_nor
```
