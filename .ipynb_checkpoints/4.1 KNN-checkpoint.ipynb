### K-近邻算法
    什么是K-近邻算法
    KNN核心思想：
        你的“邻居”来推断出你的类别


**1. K-近邻算法(KNN)原理**  

    如何确定谁是邻居？  计算距离,距离近的就是邻居  
    
    距离公式:  
        欧氏距离  
        曼哈顿距离 (绝对值距离)  
        明可夫斯基距离(欧式距离与曼哈顿距离是明科夫斯基距离的推广)  
          
**2. 注意事项:**  

    1)如果取的最近的电影数量不一样？会是什么结果？  
        k 值取得过小，容易受到异常点的影响  
        k 值取得过大，样本不均衡的影响  

    2)结合前面的约会对象数据，分析K-近邻算法需要做什么样的处理  
        无量纲化的处理  
        标准化  
                                                    

**3.优缺点**  

    优点：简单，易于理解，易于实现，无需训练  
    缺点：  
        1）必须指定K值，K值选择不当则分类精度不能保证  
        2）懒惰算法，对测试样本分类时的计算量大，内存开销大  
    使用场景：小数据场景，几千～几万样本，具体场景具体业务去测试    

**API:**  
algoright参数不用管,不会影响最终的分类结果,只是影响算法的运行效率,这是数据结构与算法层面的知识,用默认的参数就是最好的

**案例:鸢尾花分类**    
流程    
1.获取数据  
2.数据集划分
3.特征工程  
    1)特征提取(不需要)  
    2)特征预处理 --标准化  
    3)特征维度压缩(不需要)  
4.模型训练  
5.模型评估  
                      

***

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

**1.获取数据集**

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
iris.feature_names
```

```python
iris.target_names
```

**2.数据集划分**

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=22)
```

**3.特征预处理**  
**注意(非常重要):**  
对测试集,只需要transform即可,因为测试集合训练集的操作应该相同,所以做标准化时用的均值和方差应该是训练集的  
特征工程的其他步骤也是一样,对于训练集fit_transform,对于测试集,用transform()  

```python
transfer= StandardScaler()
x_train = transfer.fit_transform(x_train)
```

```python
x_test = transfer.transform(x_test)
```

**4.模型训练**

```python
estimator = KNeighborsClassifier(n_neighbors = 5)
```

```python
estimator.fit(x_train,y_train)
```

**5.模型评估**

```python
# 方法1:直接比对预测值与真实值
prediction = estimator.predict(x_test)
```

```python
prediction == y_test
```

```python
# 补充:estimator.score(x_test,y_test)算的结果和直接比对准确率的结果是一样的
# import numpy as np
# np.sum(prediction == y_test)/len(y_test)
```

```python
# 方法2:使用estimator的score方法来算准确率
estimator.score(x_test,y_test)
```
