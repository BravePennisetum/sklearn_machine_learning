```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import GridSearchCV
```

```python
# 1.获取数据
# 当用fetch直接从sklearn上下载数据时,会用到这条代码
news = fetch_20newsgroups(data_home = r'D:\codes\python_codes\machine_learning\sklearn\data',subset='all') 
```

```python
# 2.划分数据集
x_train,x_test,y_train,y_test = train_test_split(news.data,news.target,test_size=0.2)
```

```python
# 3.特征抽取
transfer = TfidfVectorizer()
x_train = transfer.fit_transform(x_train)
x_test = transfer.transform(x_test)
```

```python
# 4.模型建立
estimator = MultinomialNB()
estimator.fit(x_train,y_train)
```

```python
# 5.模型评估
estimator.score(x_test,y_test)
```
